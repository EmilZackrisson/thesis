```python
export1grechocpustat = pd.read_csv('./Testexperiment/export1/grecho1/grecho/cpu.stat.csv')
export2grechocpustat = pd.read_csv('./Testexperiment/export2/grecho2/grecho/cpu.stat.csv')

export1memorycurrent = pd.read_csv('./Testexperiment/export1/grecho1/grecho/memory.current.csv')
export2memorycurrent = pd.read_csv('./Testexperiment/export2/grecho2/grecho/memory.current.csv')

plt.plot(export1grechocpustat['snapshot'], export1grechocpustat['system_usec'].diff())
```

```python
df = pd.DataFrame({
    "time": export1grechocpustat['snapshot'],
    "system_usec": export1grechocpustat['system_usec']
})

df["dt"] = df["time"].diff()
df["rate"] = df["system_usec"].diff() / df["dt"]
df["rate_smooth"] = df["rate"].rolling(5).mean()

df.plot(x = "time", y="rate_smooth")
```
