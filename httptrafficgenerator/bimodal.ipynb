```python
import numpy as np
import matplotlib.pyplot as plt

# Generate bimodal data from two normal distributions

data1 = np.random.normal(40, 300, 400)
data2 = np.random.normal(1500, 300, 200)
data1 = data1[data1 > 0]
data2 = data2[data2 < 1500]
data = np.concatenate([data1, data2])

# Resample so the filtered dataset has exactly 1000 points
target_n = 1000
if len(data) == 0:
    raise ValueError('No samples left after filtering; cannot resample to 1000.')
replace = len(data) < target_n
data = np.random.choice(data, size=target_n, replace=replace)

# Export packet sizes for Go runtime
packet_sizes = np.clip(np.round(data), 0, 1500).astype(int)
np.savetxt('packet_sizes.txt', packet_sizes, fmt='%d')

# Create histogram
plt.hist(packet_sizes, bins=40, color='blue', edgecolor='black')
plt.title('Bimodal Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
